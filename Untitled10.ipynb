```python
# Step 1: Upload your dog or cat image
from google.colab import files
uploaded = files.upload()  # Upload a single image (e.g., cat.jpg or dog.jpg)

# Step 2: Import required libraries
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import os

# Step 3: Load the pre-trained VGG16 model
model = VGG16(weights='imagenet')

# Step 4: Set the image path (use the uploaded file name)
img_path = list(uploaded.keys())[0]

# Step 5: Load and preprocess the image
img = image.load_img(img_path, target_size=(224, 224))  # VGG16 requires 224x224
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = preprocess_input(img_array)

# Step 6: Make prediction
predictions = model.predict(img_array)
decoded = decode_predictions(predictions, top=3)[0]

# Step 7: Show predictions
print("Top predictions:")
for i, (imagenet_id, label, score) in enumerate(decoded):
    print(f"{i+1}: {label} ({score:.2f})")

# Step 8: Show image
plt.imshow(img)
plt.title('Dog vs Cat Prediction')
plt.axis('off')
plt.show()

# Step 9: Simplified decision (dog/cat based on top label)
label = decoded[0][1]
if 'cat' in label.lower():
    print("🧠 Prediction: It's a Cat 🐱")
elif 'dog' in label.lower():
    print("🧠 Prediction: It's a Dog 🐶")
else:
    print("🧠 Prediction: Not sure (Not a dog or cat)")
```
